## Explore data to determine what to include in classification model

```python
cnx = mysql.connector.connect(
    host = config.host,
    user = config.user,
    passwd = config.password,
    database = 'music')

cursor = cnx.cursor()
```

```python
## Select statement to join the song table with the features table to create a dataframe with rows containing song
## features and the artist who the song belongs to

stmt ="""
select 
        a.name, a.song_id, a.song_name, a.genre,
        music.features.feature_id, music.features.danceability, music.features.energy, music.features.key_, music.features.loudness, 
        music.features.mode, music.features.speechiness, music.features.acousticness, music.features.instrumentalness, music.features.liveness,
        music.features.valence, music.features.tempo, music.features.duration_ms, music.features.time_signature
from (
        select music.artists.name, music.artists.genre, music.songs.song_id, music.songs.song_name 
        from music.artists 
        inner join music.songs on music.artists.artist_id=music.songs.artist_id1 
        ) a
inner join
        music.features
on
        a.song_id = music.features.song_id
        
        """

cursor.execute(stmt)

songs = cursor.fetchall()
data = pd.DataFrame(songs)

# Get column names for flights dataframe
field_names = [i[0] for i in cursor.description]
```

```python
## Change columns names in dataframe to column names retrieved from SQL database

data.columns = field_names
data.columns
```

```python
data.info()
```

```python
data.describe()
```

```python
## Find out how many duplicate songs that there are based on song name. Duplicates may arise because songs may 
## appear on more than one album for an artist. For exampple, there could be a delux version of an album or 
## a compilation album for a particular artist

data[data.duplicated('song_name')==True].shape
```

```python
dropped = data.drop_duplicates('song_name', keep='first')
dropped.reset_index(drop=True, inplace = True)
```

```python
## Create bar plot that shows number of songs per artist

counts =pd.value_counts(dropped['name'])

plt.figure(figsize=(15,8))

g = sns.barplot(x=counts.index, y=counts.values)
g.set_title('Number of Songs per Artisit')
g.set_xticklabels(labels= counts.index, rotation=70)
```

## Live vs. Studio Recorded Music

### Some of these artists have live albums, which may create noise in classification of each artist

```python
live_songs = dropped[dropped['song_name'].str.contains('Live')]
live_songs
live_songs.head()
```

```python
recorded = dropped.drop(live_songs.index)
recorded.reset_index(drop=True, inplace=True)
recorded.shape
```

```python
## Jay-Z also has an acappella album which is weird, so we are going to drop that as well

a_cap = recorded[recorded['song_name'].str.contains('A Cappella')]

recorded.drop(a_cap.index, inplace=True)
```

```python
recorded.reset_index(drop=True, inplace=True)
recorded.shape
```

```python
## Create bar plot that shows number of songs per artist

counts =pd.value_counts(live_songs['name'])

plt.figure(figsize=(15,8))

g = sns.barplot(x=counts.index, y=counts.values)
g.set_title('Number of Live Songs per Artisit, Dropped for Classification')
g.set_xticklabels(labels= counts.index, rotation=70)
```

```python
counts =pd.value_counts(recorded['name'])

plt.figure(figsize=(15,8))

g = sns.barplot(x=counts.index, y=counts.values)
g.set_title('Recorded Number of Songs per Artisit')
g.set_xticklabels(labels= counts.index, rotation=70)
```

### The features for live, vs recorded music are different as seen below

```python
vars_ = ['danceability', 'energy', 'key_', 'loudness', 'speechiness', 'acousticness',
'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

for var in vars_:
    sns.distplot(live_songs[var], label='Live')
    sns.distplot(recorded[var], label='Recorded')
    plt.legend()
    plt.show()

```

## Sections Feature Engineering

```python
stmt = """select * from music.sections"""
cursor.execute(stmt)

sect = cursor.fetchall()
sections = pd.DataFrame(sect)

columns_names = [i[0] for i in cursor.description]

```

```python
sections.columns = columns_names
sections.columns
```

```python
sections.columns
keep = ['song_id', 'duration', 'loudness', 'tempo', 'key_']
sub_sections=sections[keep]
sub_sections.head()
```

```python
## Aggregate number of sections, average duration of each section, standard deviation of the duration of each section
## standard deviation of loudness for each section, and standard deviation of tempo for each song
group_sects = sections.groupby('song_id').agg({'duration': ['count', 'mean', 'std'],
                                 'loudness' : 'std',
                                 'tempo' : 'std',
                                  'key_' : pd.Series.nunique})
new_cols = ['number_sections', 'avg_duration', 'std_duration', 'std_loudness', 'std_tempo', 'number_keys']
group_sects.columns = new_cols
group_sects.head()
```

```python
group_sects.reset_index(inplace=True)
```

```python
combined = pd.merge(recorded, group_sects, left_on=recorded['song_id'], right_on=group_sects['song_id'], how='inner',)
```

```python
cols = ['name', 'song_id_x', 'song_name', 'genre', 'feature_id', 'danceability',
       'energy', 'key_', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms',
       'time_signature', 'number_sections', 'avg_duration',
       'std_duration', 'std_loudness', 'std_tempo', 'number_keys']
df = combined[cols]
```

```python
sub_cols = ['number_sections', 'avg_duration',
       'std_duration', 'std_loudness', 'std_tempo', 'number_keys']

for col in sub_cols:
    sns.boxplot(x=df['name'], y=df[col])
    plt.xlabel('')
    plt.title(col)
    plt.xticks(rotation=70)
    plt.show()
```
